```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# import seaborn as sns
%load_ext autoreload
%autoreload 2
from scipy import interpolate
from scipy import signal
import pylab
from pykalman import KalmanFilter
# from pykalman.datasets import load_robot
import time
import pickle
import math
from sklearn.decomposition import PCA, FactorAnalysis
from Bouroukan_object import spiking, cart_to_pol, rasterplot, vis_fit_AC

```

```python
# Number of neurons
ydim = 100
# linear Spike Cost
nu = 10e-2 #10e-1
# quadratic cost: distributing activity
mu = 10e-1
# Readout Decay Rate, time constant of the decoder
lbd_d = 2 # 4
# Membrane Leak, time constant of the membrane
lbd_v = 1
decoderscaling = 100#*scal/MOD.ydim*1000# **4 # 1* scal

thresh_nois_sig=.2
inp_nois_sig = 1
```

```python
# signal
ma_x = np.load('results_tmp/for Cristina/signal.npy') # format is: latent dimensions x time points x trials
dt = 1e-3 #1e-2
dtsign = 10
tmax = ma_x.shape[1]*dt
timed = np.arange(0,tmax,dt)
n_step = len(timed)
```

```python
np.random.seed(1)
SM = spiking()
SM.par(tmax, dt, timed, n_step, ydim, nu, mu, lbd_d, lbd_v, decoderscaling, J=ma_x.shape[0])

Omega, T, D = SM.connect(Dtyp='hetero') 
plt.figure()
r, th = cart_to_pol(D.T)
plt.plot(r * np.cos(th), r * np.sin(th), '-o');
plt.title('decoder')
plt.axis('square')
```

```python
X = np.zeros(ma_x.shape)*np.nan
c = np.zeros(ma_x.shape)*np.nan
dmembr = np.zeros([ma_x.shape[1], ydim])*np.nan
dspikes = np.zeros([ma_x.shape[1], ydim])*np.nan

np.random.seed(1)
X[:,:], c[:,:] = SM.signal('LDS', ma_x) # 
O, V, R, X_hat = SM.mixednet(c[:,:], D, Omega, T, seed=0, thresh_nois_sig=thresh_nois_sig)    
# translate into binned/rate form
dsamp = np.arange(0, n_step, 1/dt*dtsign)
dtim = timed[dsamp.astype(int)]
for ii in range(len(dsamp)):
    dspikes[ii,:] = np.sum(O[:,np.int(dsamp[ii]):np.int(dsamp[ii]+1/dt*dtsign-1)], axis=1)
for ii in range(len(dsamp)):
    dmembr[ii,:] = np.mean(V[:,np.int(dsamp[ii]):np.int(dsamp[ii]+1/dt*dtsign-1)], axis=1)
```

# visualize results

```python
fig, ax = plt.subplots(1,3, figsize=(15,4))
fig2, ax2 = plt.subplots(1,3, figsize=(15,4))
rasterplot(O, ax[0], tim=np.arange(0,tmax,dt))
ax[0].set_xlabel('time')
ax[1].set_xlabel('time')
ax[2].set_xlabel('time')
ax[0].set_ylabel('neuron')
r, th = cart_to_pol(D.T)
for nn in range(ydim):
    ax[1].plot(np.arange(0,tmax, dtsign), dspikes[nn,:], color=plt.cm.coolwarm(np.abs(nn-ydim/2)/(ydim/2)));
    ax[2].plot(np.arange(0,tmax, dtsign), dmembr[nn,:], color=plt.cm.coolwarm(np.abs(nn-ydim/2)/(ydim/2)));
    ax2[1].plot(r[nn] * np.cos(th[nn]), r[nn] * np.sin(th[nn]), 'o', color=plt.cm.coolwarm(np.abs(nn-ydim/2)/(ydim/2)));
    ax2[2].plot(nn, np.mean(dspikes[nn,:]), 'o', color=plt.cm.coolwarm(np.abs(nn-ydim/2)/(ydim/2)));
ax[0].set_title('spiking in trial ')
ax2[0].plot(X_hat[0,:], X_hat[1,:], 'g', label='decoded');
ax2[0].plot(ma_x[0,:], ma_x[1,:], 'k', label='signal');
ax2[1].set_aspect(1)
ax2[0].set_xlabel('dim 1')
ax2[0].set_ylabel('dim 2')
ax2[0].legend()
ax2[2].set_title('mean trial spike count')
ax2[1].set_title('deneve decoder')
ax2[0].set_title('latent')
ax[1].set_title('spike counted in bins')
ax[2].set_title('membrane aver in bins')
ax2[2].set_xlabel('neurons')
ax2[1].set_xlabel('neurons')

```
