## encoding

### adjust signal for neurons

```python
# upsample for neurons:

# Numerical Simulation parameters
dt = 1e-3
dtsign = 10
tmax = np.max(MOD.n_step)*dtsign
timed = np.arange(0,tmax,dt)
n_step = len(timed)
signscal = 10 # scal the signal in a +/- this value range to fit for the neuron
print(n_step)

newT = np.array(np.max(MOD.n_step)*dtsign/dt,dtype='int')
ma_x = np.zeros([MOD.xdim, newT, MOD.Ttrials])
# interpolate
for tt in range(MOD.Ttrials):
    f1 = interpolate.interp1d(np.arange(np.max(MOD.n_step)), MOD.x[:,0,tt], kind='cubic')
    if MOD.xdim==2: f2 = interpolate.interp1d(np.arange(np.max(MOD.n_step)), MOD.x[:,1,tt], kind='cubic')

    xnew = np.linspace(0, np.max(MOD.n_step)-1, n_step)
    if MOD.xdim==1: 
        x_ups = np.array([f1(xnew)])
    if MOD.xdim==2:
        x_ups = np.array([f1(xnew), f2(xnew)])
    
    # ... and scale
    
    for ii in range(MOD.xdim):
        ma_x[ii,:,tt] = ((x_ups[ii,:]-np.min(x_ups[ii,:]))/(np.max(x_ups[ii,:])-np.min(x_ups[ii,:]))*2-1)*signscal
if MOD.xdim==1: 
    plt.plot(xnew, ma_x[0,:,tt], 'g', label='upsampled+scal')
    plt.plot(xnew, x_ups[0,:], 'k', label='upsampled')
    plt.plot(MOD.x[:,0,tt], '.r', label='data')
if MOD.xdim==2: 
    plt.plot(ma_x[0,:,tt], ma_x[1,:,tt], 'g', label='upsampled+scaled')
    plt.plot(x_ups[0,:], x_ups[1,:], 'k', label='upsampled')
    plt.plot(MOD.x[:,0,tt], MOD.x[:,1,tt], '.g', label='data')
plt.legend()
plt.title('cubic interpolation and scaling of signal for trial '+np.str(tt))
plt.show()
```

```python
# Number of neurons
# linear Spike Cost
nu = 10e-2
# quadratic cost: distributing activity
mu = 10e-1
# Readout Decay Rate, time constant of the decoder
lbd_d = 2 # 4
# Membrane Leak, time constant of the membrane
lbd_v = 1
decoderscaling = 100#*scal/MOD.ydim*1000# **4 # 1* scal
```

```python
np.random.seed(1)
SM = spiking()
SM.par(tmax, dt, timed, n_step, MOD.ydim, nu, mu, lbd_d, lbd_v, decoderscaling, J=MOD.xdim)
X = np.zeros(ma_x.shape)*np.nan
c = np.zeros(ma_x.shape)*np.nan
dmembr = np.zeros(MOD.y.shape)*np.nan
dspikes = np.zeros(MOD.y.shape)*np.nan

Omega, T, D = SM.connect(Dtyp='hetero') # MOD.C.T) # 
plt.figure()
plt.plot(D)
```

```python
tt = 0
X[:,:,tt], c[:,:,tt] = SM.signal('LDS', ma_x[:,:,tt]) # 
O, V, R, X_hat = SM.mixednet(c[:,:,tt], D, Omega, T)    
```

```python
maxtim = 200
seed = 2 
maxiter = 50

factor = FactorAnalysis(n_components=MODtest.xdim, random_state=seed).fit(MODtest.y)
loglik = factor.score(MODtest.y)
print('FA-C PLDS likelihood %.3f' %loglik)
MODtest.estC = np.array([(factor.components_/np.max(np.abs(factor.components_)))[0]]).T # +np.random.randn(MOD.ydim)
Ctilest, dtil = MODtest.expand_C(est=1)
xtmp, ytmp = MODtest.expand_xy(est=1)
icurL = MODtest.yLik(xtmp, ytmp, Ctilest, dtil)

print(' ')
print(MODtest.ydim, ' neurons and ', MODtest.n_step, ' timepoints')
print('FA likelihood: %.3f' %icurL)
print(' ')

iiwhile = 0
conv = True
diffLik = 1
while (iiwhile<maxiter) & conv : # & (diffLik>=0)
    start = time.time()
    # ESTx0[:,iiwhile], ESTA[:,:,iiwhile], ESTQ0[:,:,iiwhile], ESTQ[:,:,iiwhile], ESTC[:,:,iiwhile],\
    #    ESTx[:,:,iiwhile], 
    _, _, _, _, _, _, conv, _, tracktime = MODtest.my_EM(ITER = 1, trackit=False, maxtimNew = maxtim, \
                                                     maxtimC = maxtim, maxtim = maxtim,\
                                                      upA = False, upQ = False, upQ0 = False)
    end = time.time()
    iiwhile +=1
    print('iteration %.1f:' %(iiwhile-1))
    Ctilest, dtil = MODtest.expand_C(est=1)
    xtmp, ytmp = MODtest.expand_xy(est=1)
    curL = MODtest.yLik(xtmp, ytmp, Ctilest, dtil)
    diffLik = (curL-icurL)
    print('improvement in log lik for y: %.3f' %diffLik)
    icurL = np.copy(curL)
    print('time for whole EM iteration: %.3f' %(end-start))
    print(tracktime)
    print(' ')
    time.sleep(.100)
print('did converge:', conv)
print(' ')
```
